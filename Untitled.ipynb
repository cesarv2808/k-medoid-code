### Algoritmo k-medoids

```python
# We import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
import random
import matplotlib.animation as animation
```

```python
# Generamos datos aleatorios
group_1 = np.abs(np.random.normal(loc=[3, 3], scale=[2.5, 2.5], size=(10, 2)))
group_2 = np.abs(np.random.normal(loc=[10, 10], scale=[2.5, 2.5], size=(10, 2)))

# Unimos los grupos
data_clustering = np.round(np.vstack([group_1, group_2]), 2)

# Creamos el daraframe
df = pd.DataFrame(data_clustering, columns=['Col1', 'Col2'])
df
```

```python
# Creamos la función que calcula la distancia manhattan
def dist(x1,x2):
    return sum(abs(a-b) for a,b in zip(x1,x2))

# creamos la funcion que obtendra los clusters junto con su costo
def clusters_(df):

    zero_cost_ = {}

    # Generamos todas combinaciones de pares de filas. Cada par corresponde a un podible medoid
    tuple_pairs = list(combinations(df.itertuples(index=False,name=None),2))
    random.shuffle(tuple_pairs) # We randomly order the list

    M = {}
    Medoids = {} # diccionario que almacenará los mdeoids junto con su costo
    cost0 = float('inf') 
    best_medoid =None
    cost1 = 0

    for pair in tuple_pairs:

        A = {}
        Bc1 = {}
        Bc2 = {}
       
        # obtenemos los primeros medoid
        c1,c2 = pair
        c1=(round(c1[0],3),round(c1[1],3))
        c2=(round(c2[0],3),round(c2[1],3))

        # iteramos sobre las filas del dataframe
        for row in df.itertuples(index=False,name=None):

            # calculamos las distancias desde cada fila a cada medoid
            dc1_ = dist(c1,row)
            dc2_ = dist(c2,row)
          
            # comparamos las distancias
            if dc1_ < dc2_:
                Bc1[row]=round(dc1_,3)
            else:
                Bc2[row]=round(dc2_,3)

        A[c1]=Bc1
        A[c2]=Bc2
        M[pair]=A

        # calculamos el costo del cluster para cada pair de medoids
        cost1 = sum(Bc1.values())+sum(Bc2.values())
        Medoids[(c1,c2)]=round(cost1,3)

    return Medoids,M
```

```python
# creamos una tabla con los resultados de los calculos anteriores

Medoids,M = clusters_(df)

# creamos las conlumnas de la tabla
columnas = {'Tupla':[],'Medoid':[],'Punto':[],'dc1':[],'dc2':[]}
columnas_ = [x for x in columnas.keys()]

# iteramos sobre el diccionario M y luego sobre los subdiccionarios contenidos en el para extraer la información y ordenarla en el diccionario columnas
for tupla,A in M.items():
    for medoid,B in A.items():
        for row,valor in B.items():
            columnas['Tupla'].append(tupla)
            columnas['Medoid'].append(medoid)
            columnas['Punto'].append(row)
            if medoid==list(A.keys())[0]:
                columnas['dc1'].append(valor)
                columnas['dc2'].append(0)
            else:
                columnas['dc1'].append(0)
                columnas['dc2'].append(valor)
                
# construimos el df con el diccionario columnas
df_=pd.DataFrame(columnas)
df_
```

```python
# construyamos grupos de clusters, en base a los pares de tuplas
# los df los guardamos en un diccionario, donde las claves serán las tuplas, es decir, los centroides de los posibles clusters
dic_groups = {columna:grupo for columna,grupo in df_.groupby('Tupla')}

# observamos la forma del primer grupo
for tuple,group in dic_groups.items():
    print(group)
    break
```

```python
# calculamos el costo para cada grupo
dic_costos = {tupla:round(sum(group['dc1'])+sum(group['dc2']),3) for tupla,group in dic_groups.items()}

# obtenemos la tupla con el menor costo
cost_min = min(x for x in dic_costos.values())
cost_min = {tupla:value for tupla,value in dic_costos.items() if value==cost_min }

# obtenemos el grupo de la tupla con el menor costo
min_group = dic_groups[list(cost_min.keys())[0]]
min_group
```
